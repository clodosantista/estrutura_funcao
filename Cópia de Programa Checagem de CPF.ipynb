```python
#Desenvolva um programa em pyhton iniciante que valide se um cpf é válido  ou não. O CPF é composto por 11 digitos e segue um padrão especifíco  de validação
def valida_cpf(cpf):
    if len(cpf) != 11:
        return False

    # Calcula e verifica primeiro dígito verificador
    soma = sum(int(cpf[i]) * (10 - i) for i in range(9))
    d1 = (soma * 10 % 11) % 10
    if d1 != int(cpf[9]):
        return False

    # Calcula e verifica segundo dígito verificador
    soma = sum(int(cpf[i]) * (11 - i) for i in range(10))
    d2 = (soma * 10 % 11) % 10
    if d2 != int(cpf[10]):
        return False

    return True

cpf = input("Digite o CPF: ")
if valida_cpf(cpf):
    print("CPF válido.")
else:
    print("CPF inválido.")
```

```python
def gerar_cpf(digitos: str) -> str:
    # Verifica se foram fornecidos 9 dígitos
    if len(digitos) != 9:
        return "Digite exatamente 9 dígitos para gerar o CPF."

    # Calcula o primeiro dígito verificador
    soma1 = sum(int(digitos[i]) * (10 - i) for i in range(9))
    resto1 = soma1 % 11
    digito1 = 0 if resto1 <= 1 else 11 - resto1

    # Calcula o segundo dígito verificador
    soma2 = sum(int(digitos[i]) * (11 - i) for i in range(9)) + digito1 * 2
    resto2 = soma2 % 11
    digito2 = 0 if resto2 <= 1 else 11 - resto2

    # Formata o CPF completo
    cpf = f"{digitos}{digito1}{digito2}"
    cpf_formatado = f"{cpf[:3]}.{cpf[3:6]}.{cpf[6:9]}-{cpf[9:]}"

    return cpf_formatado

# Exemplo de uso
nove_digitos = input("Digite os 9 primeiros dígitos do CPF: ")
cpf_gerado = gerar_cpf(nove_digitos)
print(f"CPF gerado: {cpf_gerado}")
```

```python
def gerar_digito_verificador_1(digitos):
    # Passo 1: Multiplicar cada dígito pelos pesos 10, 9, ..., 2
    pesos = [10, 9, 8, 7, 6, 5, 4, 3, 2]
    soma_multiplicacoes = sum(d * peso for d, peso in zip(digitos, pesos))

    # Passo 2: Calcular o primeiro dígito verificador
    resto_divisao = (soma_multiplicacoes * 10) % 11
    d1 = resto_divisao % 10

    return d1

def gerar_digito_verificador_2(digitos):
    # Incluindo o primeiro dígito verificador calculado anteriormente
    digitos.append(gerar_digito_verificador_1(digitos))

    # Passo 1: Multiplicar cada dígito pelos pesos 11, 10, ..., 2
    pesos = [11, 10, 9, 8, 7, 6, 5, 4, 3, 2]
    soma_multiplicacoes = sum(d * peso for d, peso in zip(digitos, pesos))

    # Passo 2: Calcular o segundo dígito verificador
    resto_divisao = (soma_multiplicacoes * 10) % 11
    d2 = resto_divisao % 10

    return d2

def gerar_cpf(cpf):
    # Converter a string de 9 dígitos para uma lista de inteiros
    digitos = [int(d) for d in cpf]

    # Calcular os dígitos verificadores
    d1 = gerar_digito_verificador_1(digitos)
    d2 = gerar_digito_verificador_2(digitos)

    # Formatar o CPF completo
    cpf_completo = f"{cpf}-{d1}{d2}"

    return cpf_completo

# Exemplo de uso:
cpf_digitos = "123456789"
cpf_gerado = gerar_cpf(cpf_digitos)
print(f"CPF gerado: {cpf_gerado}")
```
